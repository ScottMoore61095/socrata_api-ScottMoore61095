```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
endpoint = 'https://data.nashville.gov/resource/fuaa-r5cm.json'
```

```python
response = requests.get(endpoint)
```

```python
response
```

```python
result = response.json()
```

```python
search_dict = {}
for line in result:
    count = int(line.get('query_count'))
    text = line.get('query_text')
    if text in search_dict:
        search_dict[text] += count
    else:
        search_dict[text] = count
```

```python
search_df = pd.DataFrame.from_dict(search_dict, orient = 'index')
search_df = search_df.reset_index()
search_df.columns = ['term', 'text']
search_df.head()
```

```python
search_df = search_df.sort_values('text', ascending=False)
search_df.head(10)
```

```python
plt.figure(figsize=(9,6))
plt.bar('term', 'text', data = search_df.head(25))
plt.xticks(rotation = 75);
```

Question 4. Make an API request that returns the entire Top 500 Monthly Searches dataset. Make a chart that shows the number of times "maps" was searched in a month across the entire time frame.

```python
# Optional number 4. - 
# args4 = '?$q=maps&$limit=2000&$where=query_count>0'
# search = endpoint + args4
# response = requests.get(search)
# results = response.json()
# response
```

```python
args4 = '?query_text=maps&$limit=2000&$where=query_count>0'
search = endpoint + args4
response = requests.get(search)
results = response.json()
response
```

```python
search_dict = {}
for line in results:
    count = int(line.get('query_count'))
    text = line.get('month_name')
    if text in search_dict:
        search_dict[text] += count
    else:
        search_dict[text] = count
```

```python
search_df = pd.DataFrame.from_dict(search_dict, orient = 'index')
search_df = search_df.reset_index()
search_df.columns = ['query_text', 'query_count']
search_df.sort_values('query_count', ascending = False)
```

```python
plt.figure(figsize=(9,6))
plt.bar('query_text', 'query_count', data = search_df.head(25))
plt.xticks(rotation = 75);
```
